## Step 2 — Download Model to Google Drive
This downloads Qwen2.5-3B-Instruct (~6GB) directly into your Drive.
Takes ~15-20 minutes. Only ever run this once.

```python
from huggingface_hub import snapshot_download

snapshot_download(
    repo_id="Qwen/Qwen2.5-3B-Instruct",
    local_dir="/content/drive/MyDrive/Qwen2.5-3B-Instruct"
)

print("✅ Model downloaded to Google Drive!")
```

## Step 3 — Verify the Download

```python
import os

model_path = "/content/drive/MyDrive/Qwen2.5-3B-Instruct"
files = os.listdir(model_path)

print("Files in model folder:")
for f in files:
    print(f" - {f}")

# Check key files exist
required = ["config.json", "tokenizer.json", "tokenizer_config.json"]
for r in required:
    status = "✅" if r in files else "❌ MISSING"
    print(f"{status} {r}")
```
